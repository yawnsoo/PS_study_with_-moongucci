```python
today = "2022.05.19"
terms = ["A 48", "B 40", "C 3"]
privacies =["2021.05.02 A", "2021.07.01 B", "2022.02.19 C", "2022.02.20 C"]

now_date = list(map(int, today.split('.')))

dic_exp = {}

for x in terms:
    #약관 종류 dictionary로
    dic_exp[x.split(' ')[0]] = x.split(' ')[1]


my_date = []
for i in privacies:
    my_date.append(i.split(' '))

    
update_date = []


# 모든 개인정보에 대하여 각각의 만기일은 계산
for item in my_date:
    temp = list(map(int, item[0].split('.')))
    #유효기간을 몇년 몇개월로 쪼개자.
    add = int(dic_exp[item[1]])
    y = 0
    m = 0
    if add > 12:
        y = add // 12
        m = add % 12
    else:
        m = add
    tmp = temp[1]
        
    temp[0] += y
    temp[1] += m
    if temp[1] > 12:
        temp[0] += 1
        temp[1] = tmp
        temp[2] -= 1
        if temp[2] == 0:
            temp[1] -= 1
            temp[2] = 28
            if temp[1] == 0:
                temp[0] -= 1
                temp[1] = 12
            else:
                pass
        else:
            pass
    else:
        temp[2] -= 1
        if temp[2] == 0:
            temp[1] -= 1
            temp[2] = 28
            if temp[1] == 0:
                temp[0] -= 1
                temp[1] = 12

        else:
            pass
    update_date.append(temp)


answer = []
for i, item in enumerate(update_date):
    #년도 비교
    if now_date[0] > item[0]:
        answer.append(i+1)
    # 달 비교
    elif now_date[0] == item[0]:
        if now_date[1] > item[1]:
            answer.append(i+1)
        elif now_date[1] == item[1]:
            if now_date[2] > item[2]:
                answer.append(i+1)
            else:
                pass
        else:
            pass
    else:
        pass
answer
#약관 번호(index) 오름차순으로 return  이때, index 필요하므로 enumerate 활용

#40/100점

# 최대 100달 이므로 %12로 수정해야 함

# 최대 100달 해도 50점.
```
