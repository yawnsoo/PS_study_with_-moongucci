```python
park = ["OSO","OOO","OXO","OOO"]
routes = ["E 2","S 3","W 1"]
```

```python
start = [0,0] #(H, W)

for i,v in enumerate(park): #시작점 찾기
    if 'S' in v:
        start[0] = i
        for j,x in enumerate(v):
            if x == 'S':
                start[1] = j
                break
            else:
                pass
        break
    else:
        pass

W = len(park[0]) # 가로
H = len(park)    # 세로


for r in routes: # 길찾기
    h,w = start[0],start[1]
    move = int(r[-1])
    if r.startswith('E'): #동쪽
        #전체길이 체크
        if 0 <= w+move < W:
            pass
        else:
            move = 0
            continue
        #장애물 체크 
        for go in park[h][w:w+1+move]:
            if go == 'O' or go == 'S':
                continue
            else:
                move = 0
                break

        #이동
        start[1] += move
        move = 0
     
    elif r.startswith('W'): #서쪽
        
        #전체길이 체크
        if 0 <= w-move < W:
            pass
        else:
            move = 0
            continue
        #장애물 체크 
        for go in park[h][w-move:w+1]:
            if go == 'O' or go == 'S':
                continue
            else:
                move = 0
                break
        #이동
    
        start[1] -= move
        move = 0
        print(start)
    elif r.startswith('S'): #남쪽
         #전체길이 체크
        if 0 <= h+move < H:
            pass
        else:
            move = 0
            continue
        
      
        #장애물 체크 
        for go in park[h:h+1+move]:

            if go[w] == 'O' or go[w] == 'S':
                continue
            else:
                move = 0
                break

        #이동
        start[0] += move
        move = 0
   
    elif r.startswith('N'): #북쪽
        
         #전체길이 체크
        if 0 <= h-move < H:
            pass
        else:
            move = 0
            continue
        #장애물 체크 
        for go in park[h-move:h+1]:
            if go[w] == 'O' or go[w] == 'S':
                continue
            else:
                move = 0
                break

        #이동
        start[0] -= move
        move = 0
answer = start
```
